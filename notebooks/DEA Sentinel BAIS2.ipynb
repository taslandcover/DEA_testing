This notebook can be used to generate a delta - Normalised Burn Ratio image (dNBR) image for studing the effect of bushfires.

```python
# Connect to datacube containing Sentinel 2 data
dc_sentinel2 = datacube.Datacube(app='Using_load_ard')


# Temporary solution to account for Collection 3 data being in a different
# database on the NCI (is this landsat only??)
try:
    dc_landsat = datacube.Datacube(app='Using_load_ard', env='c3-samples')
except:
    dc_landsat = datacube.Datacube(app='Using_load_ard')
    
#dc = datacube.Datacube(app='load_sentinel_example')
```

```python
# Define two queries for the fire affected area, One pre-fire and one-post fire
lat = (-41.68, -41.87)
long = (145.2, 145.5)

query_pre = {
        'lat': lat,
        'lon': long,
        #'crs': 'EPSG:3577',
        'output_crs': 'EPSG:3577',
        'measurements': ['nbart_blue', 'nbart_green', 'nbart_red', 'nbart_nir_1', 'nbart_swir_3'],
        'time': ('2019-01-18'),
        'resolution': (-20, 20)
        }

query_post = {
        'lat': lat,
        'lon': long,
        #'crs': 'EPSG:3577',
        'output_crs': 'EPSG:3577',
        'measurements': ['nbart_blue', 'nbart_green', 'nbart_red', 'nbart_nir_1', 'nbart_swir_3'],
        'time': ('2019-02-27'),
        'resolution': (-20, 20)
        }
```

```python
# Load available data from both Sentinel 2 satellites
# Use regular dc_load if cloud masking is too heavy handed
ds_pre = load_ard(dc=dc_sentinel2,
              products=['s2a_ard_granule', 's2b_ard_granule'],
              **query_pre)

ds_post = load_ard(dc=dc_sentinel2,
              products=['s2a_ard_granule', 's2b_ard_granule'],
              **query_post)

# Print output data
print(ds_pre)
```

```python
'''
# Load data using dc_load (must know which sentinel is needed e.g. 'a' or 'b')
ds_pre = dc.load(product="s2a_ard_granule",
                        group_by = 'solar_day', **query_pre)
ds_post = dc.load(product='s2a_ard_granule',
                       group_by = 'solar_day', **query_post)
'''


'''
Band math version

ds_pre['NBR'] = (ds_pre.nbart_nir_1 - ds_pre.nbart_swir_3) / (dc_pre.nbart_nir_1 + ds_pre.nbart_swir_3)
ds_post['NBR'] = (ds_post.nbart_nir_1 - ds_post.nbart_swir_3) / (dc_post.nbart_nir_1 + ds_post.nbart_swir_3)
'''
```

```python
# Combine the two images into one dataset for processing
dc_comb = xr.concat([ds_pre, ds_post], dim='time')
```

```python
#Calculate NBR for the two time slices 
dc_NBR = calculate_indices(dc_comb, index='NBR', drop=True, collection='ga_s2_1')
dc_NBR
```

```python
# Plot the two timesteps side-by-side
dc_NBR.NBR.plot(vmin=-1, vmax=1, col="time")
```

```python
# Subtract pre - post to create the delta NBR image
dNBR = dc_NBR.isel(time=0) - dc_NBR.isel(time=1)
# Re-attach crs attribute
dNBR.attrs['crs'] = dc_NBR.crs
dNBR
```

```python
#Plot the delta NBR image
#dNBR
dNBR.NBR.plot(vmin=-1, vmax=1, cmap='RdYlGn')
```
