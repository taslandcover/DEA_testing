```python
query = {'x': (1385000.0, 1375000.0),
         'y': (-4570000.0, -4580000.0),
         'time': ('2016', '2018'),
         #'resolution': (50,50),
         'crs': 'EPSG:3577'}
```

```python
summer_m_pq = dc.load(product='ls8_pq_albers', fuse_func=ga_pq_fuser, group_by='solar_day',
                       **query, dask_chunks={'time': 1})

summer_m_pq = summer_m_pq.sel(time=(summer_m_pq['time.month'].isin(c_months)).dropna(dim='time'))
```

```python
summer_m_pq
```

```python
good_quality = masking.make_mask(summer_m_pq.pixelquality,
                                 cloud_acca='no_cloud',
                                 cloud_fmask='no_cloud',
                                 cloud_shadow_acca='no_cloud_shadow',
                                 cloud_shadow_fmask='no_cloud_shadow',
                                 blue_saturated=False,
                                 green_saturated=False,
                                 red_saturated=False,
                                 nir_saturated=False,
                                 swir1_saturated=False,
                                 swir2_saturated=False,
                                 contiguous=True)
```

```python
summer_m = masking.mask_invalid_data(summer_m)
summer_m = summer_m.where(good_quality)
```

```python
summer_m_gm = GeoMedian().compute(summer_m)
```

```python
# Plot geomedian layer as RGB image
summer_m_gm[['red', 'green', 'blue']].to_array().plot.imshow(robust=True)
```
