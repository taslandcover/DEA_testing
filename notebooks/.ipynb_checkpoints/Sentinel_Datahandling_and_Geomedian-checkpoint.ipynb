# Find and load Sentinel 2 Data using DEA Datahandling
This notebook uses the tools within DEADataHandling.py to find and load Sentinel data. 

*DEADatahandling is no longer supported so  use the load_ard version instead*



```python
#What can be deleted here?
%matplotlib inline
import warnings; warnings.simplefilter('ignore')

import fnmatch
import os
import pandas as pd
import geopandas as gpd

from datacube_stats.statistics import GeoMedian
from datacube.helpers import ga_pq_fuser
from datacube.storage import masking
from datacube.helpers import write_geotiff
from datacube.drivers.netcdf import write_dataset_to_netcdf
import xarray as xr

#get the DEA version of the plotting functions
import sys
sys.path.append(os.path.abspath('/g/data/r78/DPIPWE_lm/dea-notebooks/10_Scripts'))
import DEAPlotting
import DEADataHandling


import datacube
dc = datacube.Datacube(app='load_sentinel_example')
```

```python
'''Define paramaters of the data we wish to load (S2 specific)
then load data 'good' data using load_clearsentinel2.
Here the relsotion is set to 10m x 10m as the 10m S2 bands are used'''

time = ('2019-03-01')
bands = ['nbart_blue', 'nbart_green', 'nbart_red', 'nbart_nir_1']
resolution = (-10,10)

query = {
        'lat': (-40.188, -40.214),
        'lon': (148.228, 148.268),
        'output_crs': 'EPSG:3577',
        'resolution': resolution, #(-20, 20),
        'time': time #('2017-01-01', '2017-02-01')
        }


data10 = DEADataHandling.load_clearsentinel2(dc=dc, query=query,
                                                  masked_prop=0,
                                                  bands_of_interest = bands,
                                                  #mask_invalid_data = False,
                                                  mask_pixel_quality=True) 
```

```python
# Look at the 10m xarray...
data10
```

```python
'''Same command as above but this time including the SWIR_2 band (natively 20m)
and cubing at 20m x 20m'''

time = ('2018-05-25', '2018-11-15')
bands = ['nbart_blue', 'nbart_green', 'nbart_red', 'nbart_nir_1', 'nbart_swir_2']
resolution = (-20,20)

query = {
        'lat': (-40.188, -40.214),
        'lon': (148.228, 148.268),
        'output_crs': 'EPSG:3577',
        'resolution': resolution, 
        'time': time 
        }

data20 = DEADataHandling.load_clearsentinel2(dc=dc, query=query,
                                                  masked_prop=0,
                                                  bands_of_interest = bands,
                                                  #mask_invalid_data = False #Setting true alters to float64 and increases memory                                                     
                                                  mask_pixel_quality=True)
```

```python
# Look at the 20m xarray...
data20
```

```python
#Plot the first time slice of the cube for a look
DEAPlotting.three_band_image(data10, bands = ['nbart_red', 'nbart_green', 'nbart_blue'], time = 0, contrast_enhance=True);
```

# Monthly Filter and Geomedian Creation

The cells below run through the process used to filter an existing S2 datacube by month and then produce
a geomedian image. This type of process is useful where only dry season images are required e.g. for producing
dry season composite images.

```python
# function to return months of interest (cm = composite month)
def is_cm(month):
    return (month >= 11) | (month <= 4)
```

```python
# extract just the months of interest
sent_cm = data20.sel(time=is_cm(data20['time.month']))

sent_cm['time.month']
```

```python
sent_cm
```

```python
DEAPlotting.three_band_image(data20, bands = ['nbart_red', 'nbart_green', 'nbart_blue'], time = 3, contrast_enhance=True);
```

```python
# Compute geomedian here is necessary - either for dataset or subset months
sent_cmgm = GeoMedian().compute(sent_cm)
```

```python
# Plot geomedian
DEAPlotting.three_band_image(sent_cmgm, bands = ['nbart_red', 'nbart_green', 'nbart_blue'], contrast_enhance=True);
```
